# Make 1 dataset

```python
search_re = r"(0[0-9]{2}[1-9]|0[0-9][1-9][0-9]|0[1-9][0-9]{2}|1000)(?=\D|$)"

it = re.search(search_re, files[0]).group(0)
post_table = pd.read_csv(files[0],
                         skiprows=1,
                         header=None,
                         delim_whitespace=True,
                         names=headers,
                         )
iter_idx = np.array([int(it)]*55)
true_time = iter_idx * 2.5
post_table.insert(3, "iteration", iter_idx)
post_table.insert(4, "true-time", true_time)

for file in tqdm(files[1:]):
    try:
        it = re.search(search_re, file).group(1)
        last_table = pd.read_csv(file,
                                skiprows=1,
                                header=None,
                                delim_whitespace=True,
                                names=headers,
                                )
        iter_idx = np.array([int(it)]*55)
        true_time = iter_idx * 2.5
        last_table.insert(3, "iteration", iter_idx)
        last_table.insert(4, "true-time", true_time)
        post_table = pd.concat([post_table, last_table], ignore_index=True)
    except:
        print(file)
```

```python
destination = "C:/Users/enomazosii/OneDrive/document/GitHub/piml/firstProject/data/Ansys_fluent/"
post_table.to_csv(destination + "D2000_V20.csv", index=False)
```
